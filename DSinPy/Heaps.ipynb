# **Heap**
Binary tree with these two properties <br>
### **Structure property**<br>
All levels have maximum number of nodes except possibly the last level, <br>
In the last level, all the nodes are to the left<br>
hence, 
The heap tree is a complete binary tree so the height is celing of log(n+1)<br><br>

This defination is for MAX HEAP, MIN HEAPS are just opposite.
### **Heap order property**<br>
Key in any node N is greater than or equal to the keys in both children of N<br>
hence, 
root node contains the highest key


No issues with right or left, right could be greater, so be left

# Insertion in Heap
size n is incremented by 1, the new key is inserted last position of the array.<br> 
Say Key K voilates heap order property, we have to do a RestoreUp for Key K<br>
insertion here is O(h) or O(logn) 

```python
class HeapEmptyError(Exception):
    pass

class Heap:
    def __init__(self,maxSize=10):
        self.a = [None]*maxSize
        self.n = 0
        self.a[0] = 99999
    
    
    def insert(self, value):
        self.n+=1
        self.a[self.n] = value
        self.restore_up(self.n)
    
    
    def restore_up(self,i):
        k = self.a[i]
        iparent = i // 2

        while self.a[iparent] < k: # No sentinel  while(iparent>=l && self.a[iparent]<k)
            self.a[i] = self.a[iparent]
            i = iparent
            iparent = i // 2
        self.a[i] = k
    
    
    def delete_root(self):
        if self.n == 0:
            raise HeapEmptyError("Heap is empty")
        maxValue = self.a[1]
        self.a[1] = self.a[self.n]
        self.n-=1
        self.restore_down(1)
        return maxValue


```

# Deletion in Heap
 
Heap of size n <br>
key in the root is stored in some variable<br>
key in last leaf node is copied to the root node, >> key at index n is copied to index 1<br>
n - - <br>
restoreDown for key in root node <br>
same timecomplexity of O(h) or O(logn)

```python
def restore_down(self, i):
    k = self.a[i]
    lchild = 2*i
    rchild = lchild+1

    while rchild <= self.n:
        if k >= self.a[lchild] and k >= self.a[rchild]:
            self.a[i] = k
            return
        else:
            if self.a[lchild] > self.a[rchild]:
                self.a[i] = self.a[lchild]
                i = lchild
            else:
                self.a[i] = self.a[rchild]
                i = rchild
        
        lchild = 2 * i
        rchild = lchild + 1
    
    # If number of nodes is even
    if lchild == self.n and k < self.a[lchild]:
        self.a[i] = self.a[lchild]
        i = lchild
    self.a[i] = k

    
    def display(self):
        if self.n==0:
            print("heap is empty")
```
